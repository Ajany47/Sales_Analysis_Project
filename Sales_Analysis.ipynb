```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Group sales by date
daily_sales = df.groupby('Sale_Date')['Sales_Amount'].sum()
```

```python
# Plot sales trend over time
plt.figure(figsize=(12, 6))
plt.plot(daily_sales, marker='o', linestyle='-')
plt.title('Daily Sales Trend')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
df['YearMonth'] = df['Sale_Date'].dt.to_period('M')  # Convert to Year-Month format

# Group by Year-Month
monthly_sales = df.groupby('YearMonth')['Sales_Amount'].sum()

# Convert Period to string for plotting
monthly_sales.index = monthly_sales.index.astype(str)

# Plot Monthly Sales Trend
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales, marker='o', linestyle='-')
plt.title('Monthly Sales Trend')
plt.xlabel('Year-Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
# To show top 5 Highest Sales Days
```

```python
top_5_sales_days = daily_sales.sort_values(ascending=False).head(5)
print(top_5_sales_days)
```

```python
# Filter data for the highest sales date
highest_sales_df = df[df['Sale_Date'] == max_sales_date]

# Summarize key factors
top_products = highest_sales_df.groupby('Product_Category')['Sales_Amount'].sum().sort_values(ascending=False)
top_regions = highest_sales_df.groupby('Region')['Sales_Amount'].sum().sort_values(ascending=False)
top_sales_reps = highest_sales_df.groupby('Sales_Rep')['Sales_Amount'].sum().sort_values(ascending=False)
top_customers = highest_sales_df.groupby('Customer_Type')['Sales_Amount'].sum().sort_values(ascending=False)

# Display results
print("\nTop Selling Products on the Highest Sales Day:\n", top_products.head(5))
print("\nTop Selling Regions on the Highest Sales Day:\n", top_regions.head(5))
print("\nTop Sales Reps on the Highest Sales Day:\n", top_sales_reps.head(5))
print("\nTop Customer Types on the Highest Sales Day:\n", top_customers.head(5))
```

```python
# To analyse discounts and sales channel
```

```python
# Check total discount impact
total_discount = highest_sales_df['Discount'].sum()

# Sales by Payment Method
payment_methods = highest_sales_df.groupby('Payment_Method')['Sales_Amount'].sum().sort_values(ascending=False)

# Sales by Sales Channel
sales_channels = highest_sales_df.groupby('Sales_Channel')['Sales_Amount'].sum().sort_values(ascending=False)

# Display results
print(f"\nTotal Discounts Given: {total_discount}")
print("\nSales by Payment Method:\n", payment_methods)
print("\nSales by Sales Channel:\n", sales_channels)
```
